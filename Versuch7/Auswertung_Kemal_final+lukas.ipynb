```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd
import scipy
from uncertainties import ufloat
from uncertainties import umath
import math
from scipy.optimize import curve_fit
from scipy.signal import savgol_filter
```

```python
def calculateTimeIntervall(filename, peaks):
    f = open(filename, "r")
    n=float(f.readlines()[6][16:21])
    steps=[]
    for i in range(len(peaks)-1):
        steps.append(peaks[i+1]-peaks[i])
    print(n)
    return ufloat((n*np.array(steps)).mean(), (n*np.array(steps)).std())

def smallTimeIntervall(filename, peaks):
    f = open(filename, "r")
    n=float(f.readlines()[6][16:21])
    steps=[]
    for i in range(len(peaks)-1):
        steps.append(peaks[i+1]-peaks[i])
    return n*np.array(steps)
```

## Periodendauer des unged√§mpften Systems und Vgl. mit Fit

```python
filename="T0005.CSV"
df=pd.read_csv(filename, header=14)
plt.scatter(df["TIME"], df["CH1"], s =1, label="Datenpunkte")

#determine peaks
peaks=scipy.signal.find_peaks(df["CH1"], prominence=0.1, distance=10000, height=0.5)
res_period_1=calculateTimeIntervall(filename, peaks[0])
print("Period of the undamped Oscillator: " + str(res_period_1))

#make fit
def sinus_function(t, omega, phi, A):
    return A*np.sin(omega*t + phi)

popt, pcov = curve_fit(sinus_function, df["TIME"][peaks[0][0]:], df["CH1"][peaks[0][0]:], p0=[5,0,1])

fit_omega_1 = ufloat(popt[0], np.sqrt(pcov[0][0]))
fit_phaseshift_1 = ufloat(popt[1], np.sqrt(pcov[1][1]))
fit_amplitude_1 = ufloat(popt[2], np.sqrt(pcov[2][2]))
fit_period_1 = (2*np.pi)/fit_omega_1

print("omega = ", fit_omega_1)
print("Phase = ", fit_phaseshift_1)
print("Amplitude = ", fit_amplitude_1)
print("Period of the undamped Oscillator from fit:  ", fit_period_1)

# plot fit and peak
plt.scatter(df["TIME"][peaks[0][0]:], sinus_function(df["TIME"][peaks[0][0]:],
                                                     fit_omega_1.n, fit_phaseshift_1.n, fit_amplitude_1.n),s =1, label="Fitkurve")
plt.scatter(df["TIME"][peaks[0]], df["CH1"][peaks[0]], color="red", marker="^", s=8, label="Peakmaxima")

plt.xlabel(r"$t$ / s")
plt.ylabel(r"$U$ / V")
plt.legend(loc="lower left")
plt.savefig("T0_plot1.png", dpi=500, bbox_inches="tight")
plt.show()
```

get mass-weighted average values

$x_c = \left(\sum_{i=1}^N w_i \right)^{-1}\left(\sum_{i=1}^N w_i x_i \right)$ where $w_i = \dfrac{1}{\alpha_i^2}$

$\dfrac{1}{\alpha_c^2} = \left(\sum_{i=1}^N \dfrac{1}{\alpha_i^2} \right)$

```python
weight_fit = 1/((fit_period_1.s)**2) + 1/((fit_period_2.s)**2) + 1/((fit_period_3.s)**2)
error_fit = np.sqrt(1/weight_fit)
average_fit = (1/weight_fit)*( (fit_period_1.n)/((fit_period_1.s)**2) + (fit_period_2.n)/((fit_period_2.s)**2) + (fit_period_3.n)/((fit_period_3.s)**2) )
print("from fit:", average_fit, "+/-", error_fit)
```
